# 3. Rensing og inndeling

Ut ifra funnene i kvalitetssjekken så renser jeg dataene. 

### 3.1 Rensing

I kvalitetssjekken ble det funnet noe feil som nå skal rettes opp i. 

1. Fjern rader med manglende verdi i målvariabelen "retail price".
2. Fjern hele kolonnen Date, costumer postcode, store postcode, costumer x, costumer y (det ble lagt inn en strip da det skjedde noe feil under kjøring pga. costumer x og y )
3. Fjern rader i "store x" og "store y" hvor det mangler verdier.
4. Fjern duplikat
5. Gjør om datatype (da vi trenger binær til MLR)

### 3.2 Inndeling
Det deles videre inn i X (features) og y (målvariabel). Målvariabelen fjernes fra features fordi vi vil unngå at man "ser" fasiten under trening, men kun bruker den for å justere modellen etter å ha truffet/bommet. 

### 3.3 Train/test fordeling
Dataene deles i 80/20 da vi har nok data å trene på til å ikke måtte gå opp til 90/10, men det kan hende at også 70/30 kan være en fin fordeling da vi har god mengde med data.

# Felles metode for alle modellene
For å slippe gjentakelse av kode i alle modellene, lages det en felles metode for å vise resultater. 

1. MAE (Mean Absolute Error)
Finner avviket mellom prediksjonen og ekte pris, beregnet i snitt i kr.

2. MAPE
Gjør det samme som MAE, men ser på avviket i prosent i steden for kr.

3. RMSE
Denne verdien er knyttet straffer når modellen bommer. Denne verdien alene forteller oss ikke noe om resultat, men i kombinasjon med R^2 så kan det si noe om kvaliteten til modellen. Hvis RMSE er lav i forhold til prisen, så er dette bra, og motsatt i motsatt tilfelle. 

4. R^2
Denne Forklarer hvor mye variasjon i pris som forklares. Denne går i en skala fra 0 til 1, hvor nærmest 1 er best. 


Høy R^2 og lav RMSE samtidig betyr at modellen forklarer mye og bommer lite, som er det vi sikter mot. 


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluation(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)
    mape = (np.abs((y_true - y_pred) / y_true)).mean() * 100

    print("MAE  :", round(mae, 2))
    print("RMSE :", round(rmse, 2))
    print("MAPE :", round(mape, 2), "%")
    print("R^2  :", round(r2, 3))
    
    print("Snittpris (test):", round(y_true.mean(), 2))
    
```

# 5. Multippel Lineær Regressjon (MLR)

### 5.1 Trening
Vi lager en enkel LinearRegression og lar den “lære” sammenhengen mellom alle forklaringsfeltene (X_train) og prisen (y_train). Modellen finner vekter (hvor mye hvert felt drar prisen opp/ned) som passer best på treningsdata.

### 5.2 Prediksjon
Nå bruker vi den trente modellen til å gjette pris på helt nye rader vi ikke trente på, og skal få en liste med predikerte priser.

### 5.3 Resultater

MAE ligger på 35kr, som vil si at modellen bommer i snitt med 35kr per PC, som vil tilsvare MAPE på under 10%. Dette vil jeg si er ganske bra

RMSE er 50kr mens snittprisen er 489kr. Dette tilsvarer 10% som er ganske greit. - hvordan kan denne bli lavere?

R^2 ligger på 0,766 og forklarer dermed 77 % av variasjonen i pris. Dette er lavere enn jeg hadde øsnket, men kommer trolig av at det er en lineær modell. 

```python
from sklearn.linear_model import LinearRegression

# 5.1 Trening
linr = LinearRegression()
linr.fit(X_train, y_train)

# 5.2 Prediksjon
y_pred_linr = linr.predict(X_test)

# 5.3 Resultater
print("MLR :")
evaluation(y_test, y_pred_linr)
```

# 7. Random Forest (RF)


### 7.1 Trening
Koden lager mange små trær og trener dem på litt forskjellige biter av data. Hvert tre lærer sine egne enkle regler. Når alle trærne er trent, er modellen klar til å stemme på svar senere. Poenget i koden er altså: bygg mange trær, la dem lære hver for seg, og gjør dem klare til felles avstemning.

### 7.2 Prediksjoner
Et nytt punkt sendes gjennom alle trærne. Hvert tre gir et tall/snitt. Modellen tar snittet av alle tallene og det blir svaret.

### 7.3 Resultater

MAE ligger på 34 kr, som vil si at modellen bommer i snitt med 34 kr per PC, som vil tilsvare MAPE på under 10%. Dette vil jeg si er ganske bra.

RMSE er 50kr mens snittprisen er 489 kr. Dette tilsvarer ca. 10 %, som er ganske greit. 

R^2 ligger på 0,766 og forklarer dermed 77 % av variasjonen i pris. (hvorfor er den ikke høyere?)


```python
from sklearn.ensemble import RandomForestRegressor

# 7.1 Trening
rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)

# 7.2 Prediksjon
y_pred_rf = rf.predict(X_test)

# 7.3 Resultater
print("Random Forest")
evaluation(y_test, y_pred_rf)
```
